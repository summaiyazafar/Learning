```python
bins=[0,1,5,12,18,30,50,80]
labels=['infants', 'toddlers', 'kids', 'teens', 'youngs', 'middle age', 'old']
#kis columns ko bin mai add krna hai
df['binned_age']=pd.cut(df['age'], bins=bins, labels=labels)
```
