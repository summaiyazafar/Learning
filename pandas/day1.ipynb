How to rename columns:

```python
df=pd.DataFrame({"col a":[1,2,3], "col b": [4,5,6]})
df
```

```python
df.rename(columns={"col a":"col_a", "col b": "col_b"})
```

```python
df.columns=["col_AA", "col_BB"]
df
```
