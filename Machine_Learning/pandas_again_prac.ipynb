## Spliting a string into multiple columns:

```python
import pandas as pd # import lib
df= pd.DataFrame({'Name': ['Ali Hasan', 'Ahsan Khan', 'Abu Bakar', 'Zawiyar Khan'], 
                  'Location': ['Taxila Punjab', 'Lodhran Punjab', 'Karachi Sindh', 'Peshawar KPK']})# Make a dataset
df
```

```python
df[['First_Name', 'Last_Name']]= df.Name.str.split(' ', expand=True)# Spilt the dataset
df
```

```python
df[['City', 'District']]= df.Location.str.split(' ', expand=True)
df
```

```python
df= df[['First_Name', 'Last_Name', 'City', 'District']]# Refine the Manipulation
df
```
