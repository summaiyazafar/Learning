```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np

# Creating a vector v = [2, 3]
v = np.array([2, 3])

# Plotting the vector
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
