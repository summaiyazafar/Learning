## Data Visualization:

```python
#import libraries:
import pandas as pd
import numpy as sns
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import titanic dataset from seaborn library;
df= sns.load_dataset("titanic")
```

```python
df.plot(kind='scatter', x='age', y='fare')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='#8c7674')
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.head(2)
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('length Fare')
plt.xlabel('Gender(Male/Female')
```

```python
#barplot using seaborn on df:
sns.barplot(x='survived', y='fare', data=df)
```

```python
sns.barplot(data=df, x='sex', y='survived', hue='class')
```

```python
sns.boxplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.barplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.boxplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data=df,x='fare', y='age', hue='sex')
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
#make scatter plot via plotly;
import plotly.express as px
fig =px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title=dict(text='Mt Bruno Elevation'), autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
