## ` IRIS ` dataset:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.plot(kind='scatter', x='sepal_width', y='petal_width', color='#4542f5', figsize=(10,6))

```

```python

df.plot(kind='scatter', x='sepal_width', y='petal_width', color='#4542f5')
```

```python
df.groupby('petal_width')['petal_length'].mean().plot(kind='bar', color='purple', figsize=(10,4))
plt.xlabel('petal_width', rotation=360)
plt.ylabel('petal_length', rotation=90)
plt.title('Info of iris dataset')
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x='sepal_width', y='sepal_length', hue='species')
plt.show()
```

```python
import plotly.express as px
fig= px.scatter(df, x='sepal_width', y='sepal_length', color='species')
fig.show()
```
