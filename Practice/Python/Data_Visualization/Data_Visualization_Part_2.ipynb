# `Seaborn` and `Plotting`:

```python
# Data:
df=sns.load_dataset("titanic")
```

```python
df.info()
```

```python

sns.scatterplot(x='age', y='fare',data=df, hue='embark_town')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='embark_town', y='class', hue='fare')
```

```python
plt.figure(figsize=(10,5))
palette={'male':'blue', 'female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# labels:
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')

#saving a plot 300dpi, 600dpi, 150dpi png:
plt.savefig('./outputs/plots/scatter_plot.png', dpi=150)
plt.savefig('./outputs/plots/scatter_plot.svg')
```

```python
sns.set_theme(style="darkgrid", font_scale=1.2, font='Times New Roman')

plt.figure(figsize=(10,5))
palette={'male':'green', 'female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# labels:
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, hue='embark_town', x='age', y='fare' )
```

```python
plt.figure(figsize=(10,5))
plate={'male':'red', 'female':'yellow'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=plate)
```

```python
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')
plt.figure(figsize=(10,5))
info={'First':'purple', 'Second':'green', 'Third':'orange'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=info)
#labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Data of titanic passenger')
#legends
plt.legend(loc='upper right', title='Class Name')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')
plt.figure(figsize=(10,5))
info={'male':"red", 'female':'green'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='fare', palette=info)
#labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Data of titanic passenger')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')
plt.figure(figsize=(10,5))
info={'First':'red', 'Second':'green', 'Third':'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=info)
#labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Data of titanic passenger')
#legends
plt.legend(loc='upper right', title='Class Name')
#change the x-ticks name:
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=45)
```

## Barplot using `Seaborn`:

```python
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')
plt.figure(figsize=(10,5))
info={'First':'red', 'Second':'green', 'Third':'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=info, capsize=0.35)
#labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Data of titanic passenger')
#legends
plt.legend(loc='upper right', title='Class Name')
#change the x-ticks name:
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=360)
```

```python
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')
plt.figure(figsize=(10,5))
info={'First':'red', 'Second':'green', 'Third':'blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=info)
#labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Data of titanic passenger')
#legends
plt.legend(loc='upper right', title='Class Name')
#change the x-ticks name:
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=360)
```
