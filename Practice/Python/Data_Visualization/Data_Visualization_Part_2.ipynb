```python
# Data:
df=sns.load_dataset("titanic")
```

```python
df.info()
```

```python

sns.scatterplot(x='age', y='fare',data=df, hue='embark_town')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='embark_town', y='class', hue='fare')
```

```python
plt.figure(figsize=(10,5))
palette={'male':'blue', 'female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# labels:
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')

#saving a plot 300dpi, 600dpi, 150dpi png:
plt.savefig('./outputs/plots/scatter_plot.png', dpi=150)
plt.savefig('./outputs/plots/scatter_plot.svg')
```

```python
sns.set_theme(style="darkgrid", font_scale=1.2, font='Times New Roman')

plt.figure(figsize=(10,5))
palette={'male':'green', 'female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# labels:
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age and Fare')
```
